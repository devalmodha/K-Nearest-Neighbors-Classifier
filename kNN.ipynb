```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score,precision_score,recall_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
heart_df = pd.read_csv("heart.csv")
heart_df
```

```python
x = heart_df.drop(["target"],axis = 1)
y = heart_df["target"]
```

```python
y.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(
    x,y,test_size = 0.2,random_state = 42
)

scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
knn_classifiction = KNeighborsClassifier(n_neighbors=3)
knn_classifiction.fit(x_train_scaled,y_train)

y_pred = knn_classifiction.predict(x_test_scaled)

print("recall_score :",recall_score(y_test,y_pred))
print("accuracy_score :",accuracy_score(y_test,y_pred))

print("precision_score :",precision_score(y_test,y_pred))
```

```python
# k = 5

knn_classifiction = KNeighborsClassifier(n_neighbors=5)
knn_classifiction.fit(x_train_scaled,y_train)

y_pred = knn_classifiction.predict(x_test_scaled)

print("recall_score :",recall_score(y_test,y_pred))
print("accuracy_score :",accuracy_score(y_test,y_pred))

print("precision_score :",precision_score(y_test,y_pred))
```

```python
# k = 7 - best score

knn_classifiction = KNeighborsClassifier(n_neighbors=7)
knn_classifiction.fit(x_train_scaled,y_train)

y_pred = knn_classifiction.predict(x_test_scaled)

print("recall_score :",recall_score(y_test,y_pred))
print("accuracy_score :",accuracy_score(y_test,y_pred))

print("precision_score :",precision_score(y_test,y_pred))
```

```python
# k = 9
knn_classifiction = KNeighborsClassifier(n_neighbors=9)
knn_classifiction.fit(x_train_scaled,y_train)

y_pred = knn_classifiction.predict(x_test_scaled)

print("recall_score :",recall_score(y_test,y_pred))
print("accuracy_score :",accuracy_score(y_test,y_pred))

print("precision_score :",precision_score(y_test,y_pred))
```

```python
# cross validation for hyperparam tuning using griedsearch cv

from sklearn.model_selection import GridSearchCV
classifier = KNeighborsClassifier()
param_grid = {"n_neighbors" : [3,5,7,9]}

classifierCV = GridSearchCV(
    
    classifier,
    param_grid ,
    cv =5,
    scoring = "recall"
)

classifierCV.fit(x_train_scaled,y_train)
y_pred = classifierCV.predict(x_test_scaled)

print("recall_score :",recall_score(y_test,y_pred))
print("accuracy_score :",accuracy_score(y_test,y_pred))

print("precision_score :",precision_score(y_test,y_pred))


res = pd.DataFrame(classifierCV.cv_results_)
print(res[["param_n_neighbors","mean_test_score"]])

print(classifierCV.best_params_)
```

```python
# pipeline 

from sklearn.pipeline import Pipeline

x_train,x_test,y_train,y_test = train_test_split(
    x,y,test_size = 0.2,random_state = 42
)

pipeline = Pipeline([
    ('scaler', StandardScaler()), 
    ('knn', KNeighborsClassifier())
])


from sklearn.model_selection import GridSearchCV
classifier = KNeighborsClassifier()

param_grid = {"knn__n_neighbors" : [3,5,7,9]}

classifierCV = GridSearchCV(
    pipeline,
    param_grid ,
    cv =5,
    scoring = "recall"
)

classifierCV.fit(x_train,y_train)
y_pred = classifierCV.predict(x_test)

print("recall_score :",recall_score(y_test,y_pred))
print("accuracy_score :",accuracy_score(y_test,y_pred))

print("precision_score :",precision_score(y_test,y_pred))


res = pd.DataFrame(classifierCV.cv_results_)

print(classifierCV.best_params_)
```
